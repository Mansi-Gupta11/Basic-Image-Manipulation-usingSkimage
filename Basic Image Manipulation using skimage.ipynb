```python
import numpy as np
from skimage import io
from  skimage.color import rgb2gray
from skimage.util import invert
from skimage.io import imsave,imread
import matplotlib.pyplot as plt
```

```python
class Image_Manipulation():
    def __init__(self,skyline_image=None ):
        self.skyline_image = skyline_image
    def load(self):
        self.skyline_image=io.imread('download.jpg')#reading the image
        self.skyline_image[0][0]=[255,255,255]
        plt.axis("off")#removing the axis
        plt.imshow(self.skyline_image)
        print("image is loaded")
        plt.show()
        
    def edit(self):
        while True:
            try:
                option=int(input("Enter 1 for Horizontal Flip\nEnter 2 for Vertical Flip\nEnter 3 for Invert Colors\nEnter 4 for Convert to gray scale\nEnter 5 for Convert to B&W\n"))
                print()
            except:
                print("Enter the Correct Number\n")
                continue
            else:
                if option==1:
                    horizontal=self.skyline_image[::-1,:]#image is horizontally flipped
                    plt.axis("off")
                    plt.imshow(horizontal)
                    print("image is horizontall flipped")
                    plt.show()
                    
                elif option==2:
                    vertical=self.skyline_image[:, ::-1]#image is vertically flipped
                    plt.axis("off")
                    plt.imshow(vertical)
                    print("image is vertically flipped")
                    plt.show()
                    
                elif option==3:
                    inve=invert(self.skyline_image)#invert image
                    plt.axis("off")
                    plt.imshow(inve)
                    print("Colors are Inverted")
                    plt.show()
                        
                elif option==4:
                    gray_image=rgb2gray(self.skyline_image)#converting image to rgb2gray
                    plt.axis("off")
                    plt.imshow(gray_image, cmap="gray")
                    print("image is converted to grayscale")
                    plt.show()
                    
                
                elif option==5:
                    pass
                
                else:
                    exit(0)
                    break
 #driver code               
                    
    def save(self):
        io.imsave(fname="E:\GITHUB\Basic Image Manipulation\saved\sv.jpg", arr=self.skyline_image)
        print("Your Image is successfully saved")
    
obj=Image_Manipulation()
while True:
    try:
        op=int(input("Enter 1 to Load Image\nEnter 2 to Edit Image\nEnter 3 to Save Image\n"))
        print()
    except:
        print("Enter the Correct Number\n")
        continue
    else:
        if op==1:
            obj.load()
            
            
        elif op==2:
            obj.edit()
            
        elif op==3:
            obj.save()
        else:
            exit(0)
            break
        
    
    
    
```

What is skimage?

-> Skimage It is a collection of algorithms for image processing and computer vision.
-> First, we import the io module of skimage (skimage.io) so we can read and write images.
-> We import from skimage.util import invert to invert the image loaded.
-> We do from  skimage.color import rgb2gray to convert the image to grayscale


1). How to Load image?

We use the skimage.io.imread() function to read a JPEG image. Skimage reads the image, converts it from JPEG into a NumPy array, and returns the array; we save the array in a variable named skyline_image .
Next we call plt.imshow() in order to display the image.

2).How to edit the image?

-> To flip the image horizontally and vertically we used slicing.
-> To invert the image we have import "skimage.util import invert" and used the syntax  inve=invert(self.skyline_image) this will invert our image.
-> To convert to gray scale we again imported the different module from  skimage.color import rgb2gray.

-> io.imsave(fname="We Will give the path name to save file", arr="image_name")
